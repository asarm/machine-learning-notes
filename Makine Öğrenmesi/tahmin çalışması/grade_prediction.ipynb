```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split,GridSearchCV
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
df = pd.read_csv("student-mat.csv",sep=";")
df.isna().sum().sum()
```

```python
df.columns
```

```python
df.head(3)
```

```python
dms = pd.get_dummies(df[["famsize","Pstatus","internet","address","higher","activities","romantic","paid","schoolsup","famsup"]])
dms
```

```python
df = df.drop(["famsize","Pstatus","internet","address","higher","activities","romantic","paid","schoolsup","famsup"],axis=1)
```

```python
df = pd.concat([df,dms[["famsize_GT3","Pstatus_T","internet_yes","address_R","higher_yes","activities_yes","romantic_yes","schoolsup_yes","paid_yes","famsup_yes"]]],axis=1)
df.head()
```

```python
dms =  pd.get_dummies(df[["sex"]])
dms
```

```python
df = df.drop(["sex"],axis=1)
df = pd.concat([df,dms[["sex_F"]]],axis=1)
df.head(3)
```

```python
df = df.drop(["school","Mjob","Fjob","reason","guardian"],axis=1)
```

```python
dms =  pd.get_dummies(df[["nursery"]])
df = df.drop(["nursery"],axis=1)
df = pd.concat([df,dms[["nursery_yes"]]],axis=1)
df.head(3)
```

```python
#df = df.drop(["school"],axis=1)
#df.describe()
```

```python
y = df["G3"]
X = df.drop(["G3"],axis=1)
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.20,random_state=99)
```

```python
model = LinearRegression().fit(X_train,y_train)
```

```python
np.sqrt(mean_squared_error(y_test,model.predict(X_test)))
```

# SVR

```python
from sklearn.svm import SVR
svr_model = SVR("linear").fit(X_train,y_train) # linear or rbf
```

```python
y_predict = svr_model.predict(X_test)
np.sqrt(mean_squared_error(y_test,y_predict))
```

```python
pre = svr_model.predict(dataDF)
pre[0]
```

```python
svr_params = {"C":np.linspace(0.1,3,30)}
svr_cv_model = GridSearchCV(svr_model,svr_params,cv=5,n_jobs=-1).fit(X_train,y_train)
svr_cv_model.best_params_                                                                     
```

```python
svr_model = SVR("linear",C=0.1).fit(X_train,y_train)
np.sqrt(mean_squared_error(y_test,svr_model.predict(X_test)))
```
