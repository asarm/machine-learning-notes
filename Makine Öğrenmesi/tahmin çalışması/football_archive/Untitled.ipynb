```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.metrics import mean_squared_error,r2_score
import matplotlib.pyplot as plt
from sklearn.preprocessing import scale
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsRegressor
from sklearn import neighbors
from sklearn.tree import DecisionTreeRegressor
from sklearn.svm import SVR
from sklearn.neural_network import MLPRegressor
```

```python
df = pd.read_csv("football_cleaned.csv")
df = df.drop(["name","first_name"],axis=1)
df["red_card"] = df["red_card_continent"]+df["red_card_champ"]+df["red_card_cup"]
df["yellow_card"] = df["yellow_card_continent"]+df["yellow_card_champ"]+df["yellow_card_cup"]
df = df.drop(["yellow_card_cup","yellow_card_continent","yellow_card_champ","red_card_continent","red_card_champ","red_card_cup"],axis=1)
print(df.columns)
df.describe()
```

```python
df["own_goal"] = df["own_goal_continent"]+df["own_goal_champ"]+df["own_goal_cup"]
df["goal"] = df["goal_continent"]+df["goal_champ"]+df["goal_cup"]
df = df.drop(["own_goal_continent","own_goal_champ","own_goal_cup","goal_continent","goal_champ","goal_cup"],axis=1)
print(df.columns)
df.describe()
```

```python
df["clean_sheet"] = df["clean_sheet_continent"]+df["clean_sheet_champ"]+df["clean_sheet_cup"]
df["assist"] = df["assist_continent"]+df["assist_champ"]+df["assist_cup"]
df = df.drop(["clean_sheet_continent","clean_sheet_champ","clean_sheet_cup","assist_continent","assist_champ","assist_cup"],axis=1)
print(df.columns)
df.describe()
```

```python
df = df.drop(["conceded_goal_continent","conceded_goal_champ","conceded_goal_cup"],axis=1)
print(df.columns)
df.describe()
```

```python
df["penalty_goal"] = df["penalty_goal_continent"]+df["penalty_goal_champ"]+df["penalty_goal_cup"]
df = df.drop(["penalty_goal_continent","penalty_goal_champ","penalty_goal_cup","second_yellow_card_continent"],axis=1)
print(df.columns)
df.describe()
```

```python
df = df.drop(["sub_on_champ","sub_out_champ","sub_out_cup","sub_out_continent","sub_on_cup","sub_on_continent"],axis=1)
print(df.columns)
df.describe()
```
