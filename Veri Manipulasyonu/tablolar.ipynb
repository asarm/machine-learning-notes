```python
import seaborn as sns
import pandas as pd
titanic = sns.load_dataset('titanic')
titanic.head()
```

```python
titanic.groupby(['sex','class'])[['survived']].aggregate(["mean"]).unstack()
```

```python
titanic.groupby(['sex','class'])[['survived']].aggregate(["mean","sum"])
```

```python
titanic.pivot_table("survived",index = "sex",columns = ["class","adult_male"])
```

```python
titanic.groupby('sex')['fare'].mean()
```

```python
titanic.groupby('sex')['alone'].mean()
```

```python
age = pd.cut(titanic['age'],[0,25,45,90])
titanic.pivot_table("survived",["sex",age],["class"])
```
